```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Gabs = pd.read_csv('Gaborone_Temperature_Data.csv')
Global = pd.read_csv('Global_Temperature_Data.csv')
```

```python
Gaborone_Temp = Gabs['avg_temp(Gaborone)'].rolling(window=10).mean()
Global_Temp = Global['avg_temp(Global)'].rolling(window=10).mean()
```

```python
plt.plot(Gabs['year'], Gaborone_Temp, label='Gaborone_MA_temp')
plt.plot(Global['year'], Global_Temp, label='Globe_MA_temp')
plt.xlabel('year')
plt.ylabel('temp')
plt.title('temp vs year')
plt.legend()
plt.show()
```
